Import necessary libraries

```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

Upload the dataset 

```python
df = pd.read_csv('diabetes.csv')
df
```

handle the null(missing values) & duplicate values

```python
df.dropna(inplace= True)
```

```python
df.drop_duplicates(inplace= True)
```

```python
df.isnull().sum()
```

check the size of rows and columns

```python
df.shape
```

Select the Features columns and target Column

```python
X = df.drop(columns = 'Outcome')
Y = df['Outcome']
```

Split the data into train and testing data

```python
xtrain,xtest,ytrain,ytest = train_test_split(X, Y, test_size= 0.2, random_state=3)
```

train the data with different models

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
y_pred1 = lr.predict(xtest)
y_pred1
```

```python
svc = SVC(kernel= 'linear')
svc.fit(xtrain,ytrain)
y_pred2 = svc.predict(xtest)
y_pred2
```

```python
knc = KNeighborsClassifier(n_neighbors=5)
knc.fit(xtrain,ytrain)
y_pred3 = knc.predict(xtest)
y_pred3
```

```python
dtc = DecisionTreeClassifier(max_depth= 3)
dtc.fit(xtrain,ytrain)
y_pred4 = dtc.predict(xtest)
y_pred4
```

```python
rfc = RandomForestClassifier(n_estimators= 50)
rfc.fit(xtrain,ytrain)
y_pred5 =rfc.predict(xtest)
y_pred5
```

Compare the different models accuracy score

```python
acc_score1 = accuracy_score(ytest, y_pred1)
print("Accuracy score of Logistic Regressor is", acc_score1)

acc_score2 = accuracy_score(ytest, y_pred2)
print("Accuracy score of Support Vector Classifier is", acc_score2)

acc_score3 = accuracy_score(ytest, y_pred3)
print("Accuracy score of K nearest neighbor Classifier is", acc_score3)

acc_score4 = accuracy_score(ytest, y_pred4)
print("Accuracy score of Decision Tree Classifier  is", acc_score4)

acc_score5 = accuracy_score(ytest, y_pred5)
print("Accuracy score of Random Forest Classifiaction is", acc_score5)
```

Compare the different models  Cross validation Score

```python
cross_val1 = cross_val_score(LogisticRegression(), X, Y, cv= 5)
cross_val1 = cross_val1.sum()/len(cross_val1)
print("Cross validation score of Logistic Regressor is", cross_val1)

cross_val2 = cross_val_score(SVC(), X, Y, cv= 5)
cross_val2 = cross_val2.sum()/len(cross_val2)
print("Cross validation score of Support Vector Classifier is", cross_val2)

cross_val3 = cross_val_score(KNeighborsClassifier(), X, Y, cv= 5)
cross_val3 = cross_val3.sum()/len(cross_val3)
print("Cross validation score of K nearest neighbor Classifier is", cross_val3)

cross_val4 = cross_val_score(DecisionTreeClassifier(), X, Y, cv= 5)
cross_val4 = cross_val4.sum()/len(cross_val4)
print("Cross validation score of Decision Tree Classifier  is", cross_val4)

cross_val5 = cross_val_score(RandomForestClassifier(), X, Y, cv= 5)
cross_val5 = cross_val5.sum()/len(cross_val5)
print("Cross validation score of Random Forest Classifiaction is", cross_val5)
```

Different Models HyperParameter Tunning using the GridSearchCv( is used for small dataset beacuse its take too much time for executuion cuz it can work for the param_grid values are multiply with cv (like i.e of lr C = 4, max_iter= 3, cv = 5, 4*3*5 = 120 times data can train on gridsearchcv)

```python
lr = LogisticRegression()
C = [0.0001, 0.001, 0.1, 1]
max_iter = [10,100,1000]

LogisticRegressor_param_grid = {
            'C' : C,
            'max_iter' : max_iter
}

gscv1 = GridSearchCV(estimator= lr, param_grid= LogisticRegressor_param_grid, cv =5)
gscv1.fit(xtrain,ytrain)
grid_pred1 = gscv1.predict(xtest)

# check the best score and best parameter of greidsearchcv training data 

print('the gridsearchCV having the best accuracy score is', gscv1.best_score_)
print('the gridSearchCV used the best parameter is', gscv1.best_params_)
```

```python
svc = SVC()
C = [0.0001,0.001, 0.01, 0.1, 1]
kernel = ['linear', 'poly', 'rbf', 'sigmoid']

SVC_param_grid = {
            'C' : C,
            'kernel' : kernel
}

gscv2 = GridSearchCV(estimator= svc, param_grid= SVC_param_grid, cv =5)
gscv2.fit(xtrain,ytrain)
grid_pred2 = gscv2.predict(xtest)

# check the best score and best parameter of greidsearchcv training data 

print('the gridsearchCV having the best accuracy score is', gscv2.best_score_)
print('the gridSearchCV used the best parameter is', gscv2.best_params_)
```

```python
knc= KNeighborsClassifier()
n_neighbors = [5,8,10,15]

KNeighborsClassifier_param_grid = {
            'n_neighbors' : n_neighbors
}

gscv3 = GridSearchCV(estimator= knc, param_grid= KNeighborsClassifier_param_grid, cv =5)
gscv3.fit(xtrain,ytrain)
grid_pred3 = gscv3.predict(xtest)

# check the best score and best parameter of greidsearchcv training data 

print('the gridsearchCV having the best accuracy score is', gscv3.best_score_)
print('the gridSearchCV used the best parameter is', gscv3.best_params_)
```

```python
dtc = DecisionTreeClassifier()
criterion = ['gini', 'entropy']
max_depth = [2,3,4, None]
max_features = ['auto','sqrt', None]
max_leaf_nodes = [3,4,None]
random_state = [2,42,36]

DecisionTreeClassifier_param_grid = {
            'criterion' : criterion,
            'max_depth' : max_depth,
            'max_features' : max_features,
            'max_leaf_nodes' : max_leaf_nodes,
            'random_state' : random_state
}

gscv4 = GridSearchCV(estimator= dtc, param_grid= DecisionTreeClassifier_param_grid, cv =5)
gscv4.fit(xtrain,ytrain)
grid_pred4 = gscv4.predict(xtest)

# check the best score and best parameter of greidsearchcv training data 

print('the gridsearchCV having the best accuracy score is', gscv4.best_score_)
print('the gridSearchCV used the best parameter is', gscv4.best_params_)
```

It can find all the models in short lines of code using the for loop 

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# Define hyperparameters for each model
hyperparameter_of_models = {
    'LogisticRegression': {
        'C': [0.0001, 0.001, 0.1, 1],
        'max_iter': [10, 100, 1000]
    },
    'SVC': {
        'C': [0.0001, 0.001, 0.01, 0.1, 1],
        'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
    },
    'KNeighborsClassifier': {
        'n_neighbors': [5, 8, 10, 15]
    },
    'DecisionTreeClassifier': {
        'criterion': ['gini', 'entropy'],
        'max_depth': [2, 3, 4, None],
        'max_features': ['auto', 'sqrt', None],
        'max_leaf_nodes': [3, 4, None],
        'random_state': [2, 42, 36]
    },
    'RandomForestClassifier': {
        'n_estimators': [50, 100, 150, 300],
        'criterion': ['gini', 'entropy'],
        'max_depth': [2, 3, None],
        'max_features': ['auto', 'sqrt', None],
        'max_leaf_nodes': [3, 4, 5, None],
        'random_state': [2, 5, 40, None]
    }
}

# Define models
steps = {
    'LogisticRegression': LogisticRegression(),
    'SVC': SVC(),
    'KNeighborsClassifier': KNeighborsClassifier(),
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'RandomForestClassifier': RandomForestClassifier()
}


# for value, key in steps.items():
#     gscv = GridSearchCV(
#         estimator= key,
#         param_grid=hyperparameter_of_models[value],
#         cv=5,
#         n_jobs=-1  # Use -1 to utilize all available cores
#     )
#     gscv.fit(xtrain, ytrain)
#     gscv_pred = gscv.predict(xtest)

#     # Print best score and parameters for each model
#     print(f'The gridSearchCV for {value} has a best accuracy score of {gscv.best_score_}')
#     print(f'The gridSearchCV for {value} has the best parameters: {gscv.best_params_}')



# Iterate over models
for value, key in steps.items():
    rscv = RandomizedSearchCV(
        estimator= key,
        param_distributions=hyperparameter_of_models[value],
        cv=5,
        n_jobs=-1  # Use -1 to utilize all available cores
    )
    rscv.fit(xtrain, ytrain)
    rscv_pred = rscv.predict(xtest)

    # Print best score and parameters for each model
    print(f'The RandomizedSearchCV for {value} has a best accuracy score of {rscv.best_score_}')
    print(f'The RandomizedSearchCV for {value} has the best parameters: {rscv.best_params_}')
```

```python
acc_grid = accuracy_score(ytest, gscv_pred)
print('Accuracy score of grid searchCv is', acc_grid)

acc_randomized_grid = accuracy_score(ytest, rscv_pred)
print('Accuracy score of  randomizedSearchCv is', acc_randomized_grid)
```
