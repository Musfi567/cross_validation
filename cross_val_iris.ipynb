```python
X = df.drop(columns = 'Outcome')
Y = df['Outcome']
```

```python
xtrain, xtest, ytrain,ytest = train_test_split(X, Y, test_size= 0.2, stratify=Y , random_state= 2)
print(xtrain.shape, xtest.shape)
```

Multiple models that can train the data in all of these models 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
# create the list of multiple models 
models = [DecisionTreeClassifier(), LogisticRegression(),  KNeighborsClassifier(n_neighbors=5), SVC(kernel = 'linear'),RandomForestClassifier()] 
```

compare & check the accuracy score between all models

```python
for i in models:
    i.fit(xtrain, ytrain)
    y_pred = i.predict(xtest)
    accuracy = accuracy_score(ytest, y_pred)
    print('accuracy score of the all models', i,'=', accuracy)
```

compare & Check the Cross Validation Score between all models

```python
for i in models:
    cross_validataion_score = cross_val_score(i, X, Y, cv=5)
    accuracy_of_cross_val = sum(cross_validataion_score)/len(cross_validataion_score)
    print('Cross Validation score of all the models', i, '=', accuracy_of_cross_val)
```
